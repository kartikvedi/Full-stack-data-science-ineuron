```python
# Importing required libraries for visualization
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Creating a simple dataset with one feature and binary labels
X, y = make_classification(n_samples=100, n_features=1, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=42)

# Fitting Logistic Regression to the dataset
model = LogisticRegression()
model.fit(X, y)

# Generate points for plotting the decision boundary (sigmoid curve)
X_plot = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_prob = model.predict_proba(X_plot)[:, 1]  # Probability for class 1

# Plot the dataset and the sigmoid decision boundary
plt.figure(figsize=(10, 6))
plt.scatter(X, y, c=y, cmap='bwr', edgecolor='k', label="Data Points")
plt.plot(X_plot, y_prob, color='black', label='Sigmoid Curve')

# Plot vertical decision boundary (where probability = 0.5)
decision_boundary = -model.intercept_ / model.coef_[0]
plt.axvline(x=decision_boundary, color='green', linestyle='--', label=f'Decision Boundary (x = {decision_boundary[0]:.2f})')

plt.title('Logistic Regression with One Feature')
plt.xlabel('Feature Value')
plt.ylabel('Probability/Label')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Creating a dataset with two features for a more complex classification scenario
X2, y2 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

# Fitting Logistic Regression to the dataset
model2 = LogisticRegression()
model2.fit(X2, y2)

# Plotting the data points
plt.figure(figsize=(10, 6))
plt.scatter(X2[:, 0], X2[:, 1], c=y2, cmap='bwr', edgecolor='k', label="Data Points")

# Creating a meshgrid to plot the decision boundary in 2D
x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

# Predict the probabilities across the grid
Z = model2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.4, cmap='bwr')
plt.title('Logistic Regression with Two Features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Create a small dataset with two classes
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, 
                           class_sep=2, random_state=42)

# Fit the logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X, y)

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
plt.title("Generated Dataset for Logistic Regression")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Define a function to plot the decision boundary
def plot_decision_boundary(X, y, model):
    # Create a mesh grid for the plot
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    
    # Predict the probability of the mesh grid points
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot the decision boundary
    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
    plt.title("Logistic Regression Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid(True)
    plt.show()

# Plot the decision boundary
plot_decision_boundary(X, y, log_reg)
```
