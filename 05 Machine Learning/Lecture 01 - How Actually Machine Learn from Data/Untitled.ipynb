```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
x = df["Height"]
x
```

```python
y = df["Weight"]
y
```

```python
plt.title("Understand the liner regression")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.plot(x, y, c = "b", marker = "o", linestyle = "solid", label = "Original Output")

# Calculate the line using the equation: w1 = -10H + 101
predicted_W1 = [-10 * x_data + 101 for x_data in x]
plt.plot(x, predicted_W1, c = "g", marker = "o", linestyle = "solid", label = "w1 = -10H + 101")

# Calculate the line using the equation: w2 = 25.56H-82.46
predicted_W2 = [25.56 * x_data - 82.46 for x_data in x]
plt.plot(x, predicted_W2, c = "r", marker = "o", linestyle = "solid", label = "w2 = 25.56H-82.46")

# Calculate the line using the equation: w3 = 24H-72.4
predicted_W3 = [24 * x_data - 72.4 for x_data in x]
plt.plot(x, predicted_W3, c = "k", marker = "o", linestyle = "solid", label = "w3 = 24H-72.4")


plt.legend()
plt.show()
```
