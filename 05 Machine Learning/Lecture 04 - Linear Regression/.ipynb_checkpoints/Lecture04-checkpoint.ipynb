```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Advertising.csv")
df
```

```python
data_X = df[["TV", "Radio", "Newspaper"]]
data_X
```

```python
y = df["Sales"]
y
```

```python
model = LinearRegression()
```

```python
model.fit(data_X, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#equation looks like
#y = (0.04576465*Tv + 0.18853002*Radio -0.00103749*Newspaper) + 2.9
#y = m1x1 + m2x2 + m3x3 + c
```

```python
model.score(data_X, y) #confident score
```

```python
#using 2 columns build another model
data_X = df[["TV", "Radio"]]
y = df.Sales
```

```python
model1 = LinearRegression()
model1.fit(data_X, y)
model1.score(data_X, y)
```

```python
import statsmodels.formula.api as smf
model2 = smf.ols(formula = "Sales ~ TV", data = df).fit()
model2.summary()
```

```python
model2 = smf.ols(formula = "Sales ~ TV + Radio", data = df).fit()
model2.summary()
```

```python
model2 = smf.ols(formula = "Sales ~ TV + Radio + Newspaper", data = df).fit()
model2.summary()
```

```python
model2 = smf.ols(formula = "Sales ~ Radio + Newspaper", data = df).fit()
model2.summary()
```

```python
model2 = smf.ols(formula = "Sales ~ Newspaper", data = df).fit()
model2.summary()
```
